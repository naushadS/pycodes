```python
from IPython.display import clear_output
from huepy import *
from itertools import combinations
from random import randint

class TicTacToe:
    
    def __init__(self):
        self.board = list(map(lambda x:grey(x), list(range(1,10))))
        self.points_marked = []
        self.player1 = None
        self.player2 = None
        self.player1_mark = None
        self.player2_mark = None
        
    def print_board(self):
        print('{0} | {1} | {2}\n~~~~~~~~~\n{3} | {4} | {5}\n~~~~~~~~~\n{6} | {7} | {8}'.format(*self.board))
        
    def reset_board(self):
        self.board = list(map(lambda x:grey(x), list(range(1,10))))
        self.points_marked = []
        self.player1_mark = None
        self.player2_mark = None
        
    def mark_on_board(self,point,entry):
        if(point in self.points_marked or (point==0 or point>9)):
            if point in self.points_marked:
                print('Already Marked ! Choose another unmarked space on board.') 
            if entry==self.player1_mark:
                return self.mark_on_board(int(input(self.player1 + "'s turn:" + self.player1_mark + " your move: ")),self.player1_mark) 
            else:
                return self.mark_on_board(int(input(self.player2 + "'s turn:" + self.player2_mark + " your move: ")),self.player2_mark)
            
        if(point>=1 and point<=9):
            self.board[point-1] = red(entry) if entry=='X' else blue(entry)
            self.points_marked.append(point)
        
        clear_output()
        self.print_board()
        
        # For a winning match, there should be atleast a total of 5 points marked by both players on the board
        # So no point in checking winning match before that
        if(len(self.points_marked)<5):
            return False
        
        return entry if self.check_match(entry) else False
        
    def check_match(self,entry):
        horizonal_match = [(0,1,2),(3,4,5),(6,7,8)]
        vertical_match = [(0,3,6),(1,4,7),(2,5,8)]
        diagonal_match = [(0,4,8),(2,4,6)]
        winning_match = horizonal_match + vertical_match + diagonal_match
        
        entry = red(entry) if entry=='X' else blue(entry)
        marked_loc_indices = [index for index,item in enumerate(self.board) if item==entry]
        
        if (len(marked_loc_indices)<3):
            return False
        
        marked_loc_indices.sort()
        size_3_marked_loc_indices = list(combinations(marked_loc_indices, 3))
        
        for tup in size_3_marked_loc_indices:
            if tup in winning_match:
                return True
        else:
            return False
        
    def replay_option(self):
        self.replay = None
        while self.replay != 'y' and self.replay != 'n':
            self.replay = input("Do you want to replay? (y or n) : ").lower()
        if self.replay=='y':
            self.reset_board()
            clear_output()
            self.start()
        else:
            clear_output()
            print('Thank you for playing!')
    
    def play(self):
        self.print_board()
        print('Enter the number where you want to mark')
        while True:
            if self.first_to_play==0:
                if (len(self.points_marked)!=9 and self.mark_on_board(int(input(self.player1 + "'s turn:" + self.player1_mark + " your move: ")),self.player1_mark)==self.player1_mark):
                    print(red(self.player1 + ' ') + green('won !')) if self.player1_mark == 'X' else print(blue(self.player1 + ' ') + green('won !'))
                    break

                if (len(self.points_marked)!=9 and self.mark_on_board(int(input(self.player2 + "'s turn:" + self.player2_mark + " your move: ")),self.player2_mark)==self.player2_mark):
                    print(red(self.player2 + ' ') + green('won !')) if self.player2_mark == 'X' else print(blue(self.player2 + ' ') + green('won !'))
                    break
            else:
                if (len(self.points_marked)!=9 and self.mark_on_board(int(input(self.player2 + "'s turn:" + self.player2_mark + " your move: ")),self.player2_mark)==self.player2_mark):
                    print(red(self.player2 + ' ') + green('won !')) if self.player2_mark == 'X' else print(blue(self.player2 + ' ') + green('won !'))
                    break
                    
                if (len(self.points_marked)!=9 and self.mark_on_board(int(input(self.player1 + "'s turn:" + self.player1_mark + " your move: ")),self.player1_mark)==self.player1_mark):
                    print(red(self.player1 + ' ') + green('won !')) if self.player1_mark == 'X' else print(blue(self.player1 + ' ') + green('won !'))
                    break
            
            if (len(self.points_marked)==9):
                print('~~~~~~~~~~~~~~~~~~~~DRAW~~~~~~~~~~~~~~~~~~')
                break
        self.replay_option()
                
    def mark_selection(self):
        while self.player1_mark != 'X' and self.player1_mark != 'O':
            self.player1_mark = input(self.player1 + ", Choose your mark! (X or O) :").upper()
        if self.player1_mark=='X':
            self.player2_mark='O' 
        else:
            self.player2_mark='X'
        
    def start(self):
        if self.player1 == None and self.player2 == None:
            self.player1 = input("Player 1's name: ")
            self.player2 = input("Player 2's name: ")
        self.mark_selection()
        ready_to_play = None
        while ready_to_play != 'y':
            ready_to_play = input("Are you ready to play? (y or n) :").lower()
        self.first_to_play = randint(0,1)
        print(self.player1 + " will go first!") if self.first_to_play==0 else print(self.player2 + " will go first!")
        self.play()
```
